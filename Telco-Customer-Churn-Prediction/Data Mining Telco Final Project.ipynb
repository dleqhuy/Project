```python
dataset = df.copy()
dataset['Churn'].replace(to_replace='Yes', value=0, inplace=True)
dataset['Churn'].replace(to_replace='No',  value=1, inplace=True)

X = dataset.drop(columns = ['customerID','Churn'])
y = dataset.Churn
```

```python
cols_cat = list(X.select_dtypes(include = "object").columns)
```

```python
cols_num = list(X.select_dtypes(exclude = "object").columns)
```

```python
print("Các cột categorical trong dữ liệu là:\n {}".format(cols_cat))
```

```python
print("Các cột numerical trong dữ liệu là:\n {}".format(cols_num))
```

Thực hiện one hot encoding cho các tính năng phân loại có trong bộ dữ liệu.

```python
df_dummy_encoding = pd.get_dummies(X[cols_cat], drop_first = True)
```

Combining the numerical and categorical features to form the final dataframe of interest. 

```python
X_final = pd.concat([df_dummy_encoding, X[cols_num]], axis = 1)
```
